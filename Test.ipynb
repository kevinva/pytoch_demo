```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def x_function(x):
    return 3 * (x ** 2) - 60 * x + 9
```

```python
x_list = [i for i in np.linspace(0, 100)]
y_list = [x_function(x) for x in x_list]
plt.plot(x_list, y_list)
plt.show()
```

```python
# 初始温度
T = 1

# 随机初始化一个解
x = np.random.uniform(0, 100)

# 终止温度
std = 0.00000001

# 衰减率
a = 0.99

loss_list = []
while T > std:
    y = x_function(x)
    x_new = x + np.random.uniform(-1, 1)
    if x_new >= 0 and x_new <= 100:
        y_new = x_function(x_new)
        loss = y - y_new
        if loss > 0:
            x = x_new
        else:
            p = np.exp((y - y_new) / T)
            r = np.random.uniform(0, 1)
            if p > r:
                x = x_new
        
        loss_list.append(loss)
                
    T = T * a
    
print(x, x_function(x))
```

```python
xs = range(0, len(loss_list))
plt.plot(xs, loss_list)
plt.show()
```
